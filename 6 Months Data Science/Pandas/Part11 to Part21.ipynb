```python
#Function of batsmen who scored maximum run against his opponent
def run_scored(batsman_name):
    vk =delivery[delivery['batsman']== batsman_name]
    return vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending = False).index[0]
    
    
```

```python
run_scored('G Gambhir')
```

```python
#isin()
# Find the most destructive death over batsman in history of IPL
# Strike rate = (Number of runs/Number of balls)/100
#Min batsman 200 balls in 16-20 over
```

```python
delivery['over']>15
```

```python
death_over = delivery[delivery['over']>15]
death_over
```

```python
 all_bat = death_over.groupby('batsman')['batsman_runs'].count().sort_values(ascending = False)
 x =all_bat>200
 #this is masking
 all_bat[x]
```

```python
#Extracting index from series
all_bat[x].index
```

```python
#Convert index to list

batsman_list = all_bat[x].index.tolist()
batsman_list
```

```python
#Runs scored by all these 43 batsman
#Balls played by these 43 batsman

#Now we will use isin for multiple masking
final = delivery[delivery['batsman'].isin(batsman_list)]
```

```python
runs = final.groupby('batsman')['batsman_runs'].sum()
```

```python
balls = final.groupby('batsman')['batsman_runs'].count()
```

```python
SR = (runs/balls)*100
SR.sort_values(ascending=False)
```

# Handling Missing Values

1. Drop Missing Data using dropna()

```python
# we will perform all this operation on titanic datasets
data = pd.read_csv('train.csv')
data.head()
```

```python
data.shape
```

```python
data.dropna().shape
```

```python
#hows only remove that remove which does not any value
data.dropna(how='all').shape
```

```python
#it remove those who have only just 1 missing value
data.dropna(axis = 1).shape
```

```python
#it remove those who have only those column who all values are null

data.dropna(axis = 1,how = 'all').shape
```

```python
data.dropna(subset=['Cabin']).shape
```

```python
#there is only 2 place where embarked value is missing, total data is 891 and embarked is 889
data.dropna(subset=['Embarked']).shape
```

```python
#by doing multiple columns it considered in union/OR
data.dropna(subset=['Embarked','Cabin']).shape
```
