# BoxPlots

A boxplots is a standardized way of displaying the distribution of data based on a five number summary("minimum", first quartile(Q1), median, third quartile(q3) and "maximum"). It can tell you about your outliers and what their values are. It
can also tell if your data is symmetrical, how tightly your data is grouped, and if and how your data is skewed.

```python


import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.describecribe()
```

```python
#univariate

sns.boxplot(tips['tip'])
```

```python
#univariate

sns.boxplot(tips['total_bill'])
```

```python
#multivariant

sns.catplot(x = 'day', y = 'total_bill', kind = 'box', data = tips)
```

```python
sns.catplot(x = 'day', y = 'total_bill', hue = 'sex', kind = 'box', data = tips)
```

# VILOIN PLOTS

violin plots are method of plotting numeric data and can be considered a combination of box plot with a kernal density plot

1. The advantage of violin plot over box plot is that aside from showing the above mentioned statistics, it also shows the entire distribution of the data
2. Violin plots are often used to compare the distribution of the given variable across some categories

```python
sns.violinplot(tips['total_bill'])
```

```python
sns.catplot(x = 'day', y = 'total_bill', kind = 'violin', data = tips)
```

```python
sns.catplot(x = 'day', y = 'total_bill', kind = 'violin', hue = 'sex', data = tips)
```

```python
sns.catplot(x = 'day', y = 'total_bill', kind = 'violin', hue = 'sex',  split = True, data = tips)
```
