# Types of Unclean Data

## There are 2 kind of unclean Data

1. Dirty Data(Data with Quality issues):
    Dirty data, also known as low quality data. Low quality data has content issues

2. Messy Data( Data with tidiness issues):
    Messy Data, also known as untidy data. Untidy data has structural issues.

# 1.Dirty Data

## When the Data has following of isssues:-
    1. Duplicated Data
    2. Missing Data
    3. Corrupt Data
    4. Inaccurate Data
    
This kind of data is known as Dirty Data

# 2. Messy Data

##  When the Data has structural issue is it known as Messy or Untidy Data.

Tidy Data has the following properties:
    
    1. Each variable forms a column
    2. Each observation forms a row
    3. Each observation unit forms a table

if a data violates any one of above it is known as untidy data

## Cleaning Data

You should follow the order provided below:-
   - Completeness Issues
   - Tidiness Issues
   - Remaining Data Quality Issues like validity, Accuracy and Consistency

# Steps Involve in Data Cleaning

- Define
- Code
- Test

**Always make sure to create a copy of your pandas dataframe before you start the cleaning proces**


```python
#Creating copies

Patients_copy = patients.copy()
treatments_copy = treatments.copy()
advers_copy = advers.copy()
```

```python
Patients_copy[Patients_copy['address'].isnull()]
```

###  a. Completeness Issues

- Missing values for contact details of 12 patients
- Only 280 rows are there, instead of 350
- hba1c_change column has missing values



#### Define

```python
# Subtract hba1c_start from hba1c_end to get hba1c_Change
```

#### Code

```python
treatments_copy['hba1c_Change'] = treatments_copy['hba1c_start'] - treatments_copy['hba1c_end']
```

#### Test

```python
treatments_copy.isnull().sum()
```

```python
treatments_copy.describe()
```

###  b. Tideness Issue

- Phone + email in contact column patients
- Novodra and Aurlin column treatment
- Merge the adverse table to treatment table



#### Define

```python
# Novodra and Aurolin using the melt function
```

#### Code

```python
treatments_copy = pd.melt(treatments_copy, id_vars=['given_name', 'surname', 'hba1c_start', 'hba1c_end', 'hba1c_change'],
       var_name='treatment', value_name='dose')
```

```python
treatments_copy
```

```python
treatments_copy = treatments_copy[treatments_copy['dose']!='-']
```

```python
treatments_copy['start'],treatments_copy['end'] = treatments_copy['dose'].str.split('-').str
```

```python
treatments_copy
```

```python
treatments_copy.drop(columns = {'dose'}, inplace= True)
```
